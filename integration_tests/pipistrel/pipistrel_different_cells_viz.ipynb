```python
MISSION_DATA_FILE = pth.join(RESULTS_FOLDER_PATH, "mission_data_high_lifespan_cell.csv")
PT_DATA_FILE = pth.join(
    RESULTS2_FOLDER_PATH, "pipistrel_power_train_data_with_lca_reference_cell.csv"
)

perfo_viewer = oad_he.PerformancesViewer(
    power_train_data_file_path=PT_DATA_FILE,
    mission_data_file_path=MISSION_DATA_FILE,
    plot_height=800,
)
```
